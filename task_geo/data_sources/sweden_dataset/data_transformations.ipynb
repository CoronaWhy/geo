# Disposable income

```python
read_str = 'data_sources/000000KD.csv'
df = pd.read_csv(read_str, encoding='cp1252')
```

```python
df.rename(columns={"Mean value, SEK thousands 2018": "mean", "Median value, SEK thousands 2018": "median"}, inplace=True)
df.replace(' ', '_', regex=True, inplace=True)
```

```python
df = pd.melt(df, id_vars=['region', 'age'], value_vars=['mean', 'median'])
```

```python
df['age_value'] = df[['age', 'variable']].agg('_'.join, axis=1)
df.drop(['age', 'variable'], axis=1, inplace=True)
```

```python
df_out = df.pivot_table(values='value', index=['region'], columns='age_value')
df_out.reset_index(inplace=True)

df_out
```

```python
df_out.to_csv('income.csv', encoding='utf-8', index=False)
```
